```python
def crop_video_into_video(file_path):
  face_locations = []
  face_encodings = []
  face_names = []
  process_this_frame = True
  frame_store = []#전체 프레임이 저장될 영역
  frame_count = 0
  min_top = 99999
  max_right = 0
  max_bottom = 0
  min_left = 99999
  video_capture = cv2.VideoCapture(file_path)#파일 경로로부터 비디오 읽어옴.
  while True:
      # Grab a single frame of video
      ret, frame = video_capture.read()
      if(ret == False):
        break
      frame_store.append(frame)
      #frame_count+=1
      # Resize frame of video to 1/4 size for faster face recognition processing
      small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)
      # Convert the image from BGR color (which OpenCV uses) to RGB color (which face_recognition uses)
      rgb_small_frame = small_frame[:, :, ::-1]

      # Only process every other frame of video to save time
      if process_this_frame:
          face_locations = face_recognition.face_locations(rgb_small_frame)
          face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)

      # Display the results
      for (top, right, bottom, left) in face_locations:
          # Scale back up face locations since the frame we detected in was scaled to 1/4 size
          top *= 4
          right *= 4
          bottom *= 4
          left *= 4
          
          min_top = min(min_top, top)
          max_right = max(right, max_right)
          max_bottom = max(bottom, max_bottom)
          min_left = min(left, min_left)
  frame_store = [x for x in frame_store if x is not None]
  return frame_store,min_top, max_right, max_bottom, min_left
```

```python
celeb_df_path = cur_path+'/celeb-df_v2'
print(celeb_df_path)
celeb_df_path_list = os.listdir(celeb_df_path)
print(celeb_df_path_list)
out_path_origin = cur_path+'/celeb-df_v2/pre_process'

total = 0
for i in celeb_df_path_list:
  if('Celeb-synthesis' in i):
    continue
  if('.txt' not in i):
    inside_folder_path = celeb_df_path+'/'+i
    inside_list = os.listdir(inside_folder_path)
    total = len(inside_list)
    count = 0
    for k in inside_list:
      count+=1
      if('.mp4' in k):
        #print(inside_folder_path+'/'+k)
        out_path = out_path_origin+'/'+i+'/'+k.split('.')[0]+'.avi'
        #print(out_path)
        if(os.path.isfile(out_path)==True):
          #print('already finished : '+out_path)
          continue
        else:
          #print('미완료 작업 시작:'+k)
          frame_list, min_top, max_right, max_bottom, min_left = crop_video_into_video(inside_folder_path+'/'+k)
          #print((len(frame_list)),min_top, max_right, max_bottom, min_left)
          size = (max_right-min_left, max_bottom-min_top)
          fps = 30.0
          #print(out_path)
          out = cv2.VideoWriter(out_path,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
          if not out.isOpened():
            print(inside_folder_path+'/'+k)
            print('File open failed!')
          for j in range(len(frame_list)):
            out.write(frame_list[j][min_top:max_bottom, min_left:max_right])
          out.release()
        print('current/total :',str(count)+'/'+str(total))
```
